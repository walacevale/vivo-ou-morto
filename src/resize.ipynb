```python
import matplotlib.pyplot as plt
import cv2
import os
import glob
```

```python
def resize(img,size,name, caminho):
    img_color = cv2.cvtColor(img, cv2.COLOR_BGR2RGB )
    img_resize = cv2.resize(img_color, (size,size))
    plt.imsave(caminho + str(name) + '.png', img_resize )
    return
```

```python
pastas_img = ['train' , 'test']
caminho_base = os.path.abspath('..')

def caminho_image(numero_pasta, formato):
    caminho_pastas_imgs = caminho_base + '/image' + '/trypoastigote_images/' +str(pastas_img[numero_pasta]) + '/'
    caminho_imgs = glob.glob(caminho_pastas_imgs + '*.' + str(formato))
    return caminho_imgs

def caminho_image_save(numero_pasta, formato):
    caminho_pastas_imgs = caminho_base + '/image/' + '/trypoastigote_images_resize/' +str(pastas_img[numero_pasta]) + '/'
    
    return caminho_pastas_imgs
```

```python
SIZE = 1024
caminho_try  = caminho_image(0, 'jpg')
for varrer in range(len(caminho_try)):
    image = plt.imread(caminho_try[varrer])
    resize(image, SIZE, varrer, caminho_image_save(0,'png'))
    
```
